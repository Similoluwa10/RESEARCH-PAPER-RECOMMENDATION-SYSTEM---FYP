## Generate Explanation

```python
# Example recommendation
query = "deep learning for software testing"
recommended_title = "Neural Network Approaches for Automated Test Case Generation"

explanation = explainer.generate(
    query=query,
    recommended_title=recommended_title,
    similarity_score=0.82,
    key_terms=["deep learning", "neural network", "testing", "automated"],
    similarity_breakdown={
        "semantic": 0.85,
        "keyword": 0.72,
        "overall": 0.82,
    },
)

print("Explanation:")
print(f"  Summary: {explanation['summary']}")
print(f"  Key Terms: {explanation['key_terms']}")
print(f"  Confidence: {explanation['confidence']}")
```
