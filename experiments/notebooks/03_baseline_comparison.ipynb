# Baseline Comparison

This notebook compares semantic search against baseline methods.

## Methods Compared
1. Semantic Search (Sentence Transformers)
2. TF-IDF
3. BM25
4. Keyword Matching

```python
import sys
sys.path.insert(0, '../../packages/nlp')
sys.path.insert(0, '../evaluation')

from src.baselines import TFIDFBaseline, BM25Baseline
from src.embeddings import SentenceTransformerEmbedding
from benchmark import BenchmarkRunner
```

```python
# Sample corpus
corpus = [
    "Deep learning approaches for automated code review",
    "Machine learning based bug detection in software",
    "Neural network models for code completion",
    "Traditional static analysis for vulnerability detection",
    "Empirical study of code review practices",
]

query = "machine learning for code analysis"
```

## TF-IDF Baseline

```python
tfidf = TFIDFBaseline()
tfidf.fit(corpus)

tfidf_results = tfidf.search(query, top_k=3)
print("TF-IDF Results:")
for idx, score in tfidf_results:
    print(f"  [{score:.4f}] {corpus[idx]}")
```

## BM25 Baseline

```python
bm25 = BM25Baseline()
bm25.fit(corpus)

bm25_results = bm25.search(query, top_k=3)
print("BM25 Results:")
for idx, score in bm25_results:
    print(f"  [{score:.4f}] {corpus[idx]}")
```
